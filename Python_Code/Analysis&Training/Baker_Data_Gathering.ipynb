```python
def singel_game_rush_stats(x,team):
    game_rush=pd.read_html(x)[0]
    game_rush=game_rush[['Unnamed: 0_level_0','Unnamed: 1_level_0','Rushing']]
    game_rush=game_rush.set_index(game_rush['Unnamed: 1_level_0']['Tm'])
    game_rush=game_rush.reset_index()
    game_rush.drop(['Unnamed: 1_level_0'], axis=1, inplace=True)
    game_rush=game_rush.set_index(game_rush['Unnamed: 0_level_0']['Player'])
    game_rush=game_rush[game_rush['Tm']==team]
    game_rush.drop(['Tm'], axis=1, inplace=True)
    game_rush=game_rush['Rushing'].reset_index()
    game_rush['Yds']=game_rush['Yds'].astype(int)
    game_rush=game_rush[game_rush['Yds']> 0]
    game_rush['Att']=game_rush['Att'].astype(int)
    game_rush['Yds']=game_rush['Yds'].astype(int)
    game_rush['TD']=game_rush['TD'].astype(int)
    game_rush['Lng']=game_rush['Lng'].astype(int)
    game_rush['Y/A']=game_rush['Yds'].astype(int)/game_rush['Att'].astype(int)
    game_rush=game_rush[['Player', 'Att', 'Yds', 'Y/A', 'TD', 'Lng']]
    game_rush['Year']=2022
    return game_rush
```

```python
def home_game(x,y):
    if x['Tm'][0]==y:
      return 0
    else:
      return 1
```

```python
def oppont(x,y):
    if x['Tm'][0]==y:
      return x['Tm'].unique()[1]
    else:
      return x['Tm'].unique()[0]
```

```python
def passing_performance(x,team,year):
    baker_pass=pd.read_html(x)[0]
    baker_pass=baker_pass[['Unnamed: 0_level_0','Unnamed: 1_level_0', 'Passing']]
    baker_pass=baker_pass.set_index(baker_pass['Unnamed: 1_level_0']['Tm'])
    baker_pass=baker_pass.reset_index()
    baker_pass.drop(['Unnamed: 1_level_0'], axis=1, inplace=True)
    baker_pass=baker_pass.set_index(baker_pass['Unnamed: 0_level_0']['Player'])
    baker_pass=baker_pass.reset_index()
    baker_pass.drop(['Unnamed: 0_level_0'], axis=1, inplace=True)
    baker_pass=baker_pass.set_index(['Player','Tm'])
    baker_pass=baker_pass['Passing'][['Cmp','Att','Yds','TD','Int','Sk','Yds.1','Lng','Rate']]
    baker_pass=baker_pass.reset_index()
    baker_pass['Home Game']=home_game(baker_pass, team)
    baker_pass['Opponent']=oppont(baker_pass,team)
    baker_pass['Season']=year
    baker_pass=baker_pass[baker_pass['Tm']==team]
    baker_pass=baker_pass.dropna()
    baker_pass=baker_pass[baker_pass['Att']==baker_pass['Att'].max()]
    return baker_pass
```

```python
def singel_game_wr_rec_stats(x,team,year):
    game_1=pd.read_html(x)[0]
    game_1=game_1[['Unnamed: 0_level_0','Unnamed: 1_level_0','Receiving']]
    game_1=game_1.set_index(game_1['Unnamed: 1_level_0']['Tm'])
    game_1=game_1.reset_index()
    game_1.drop(['Unnamed: 1_level_0'], axis=1, inplace=True)
    game_1=game_1.set_index(game_1['Unnamed: 0_level_0']['Player'])
    game_1['Home Game']=home_game(game_1,team)
    game_1['Opponent']=oppont(game_1,team)
    game_1=game_1[game_1['Tm']==team]
    game_1.drop(['Unnamed: 0_level_0'], axis=1, inplace=True)
    game_2=game_1['Receiving'].reset_index()
    game_2=pd.merge(game_2,game_1, how='left', on='Player')
    game_2=game_2.iloc[:,[0,1,2,3,4,5,6,12,13]]
    game_2['Tm']=game_2.iloc[:,6]
    game_2['Home Game']=game_2.iloc[:,7]
    game_2['Opponent']=game_2.iloc[:,8]
    game_2=game_2.iloc[:,[0,1,2,3,4,5,9,10,11]]
    game_2=game_2[game_2['Yds']> 0]
    game_2['Y/R']=round(game_2['Yds']/game_2['Rec'],1)
    game_2=game_2.iloc[:,[0,1,2,3,9,4,5,6,7,8]]
    game_2['Year']=year
    return game_2
```

```python
def singel_game_rush_stats(x,team,year):
    game_rush=pd.read_html(x)[0]
    game_rush=game_rush[['Unnamed: 0_level_0','Unnamed: 1_level_0','Rushing']]
    game_rush=game_rush.set_index(game_rush['Unnamed: 1_level_0']['Tm'])
    game_rush=game_rush.reset_index()
    game_rush.drop(['Unnamed: 1_level_0'], axis=1, inplace=True)
    game_rush=game_rush.set_index(game_rush['Unnamed: 0_level_0']['Player'])
    game_rush['Home Game']=home_game(game_rush,team)
    game_rush['Opponent']=oppont(game_rush,team)
    game_rush=game_rush[game_rush['Tm']==team]
    game_rush.drop(['Unnamed: 0_level_0'], axis=1, inplace=True)
    game_rush_2=game_rush['Rushing'].reset_index()
    game_rush_2=pd.merge(game_rush_2,game_rush, how='left', on='Player')
    game_rush_2=game_rush_2.iloc[:,[0,5,1,2,3,4,6,7,8,9,10,11]]
    game_rush_2=game_rush_2[game_rush_2['Yds']> 0]
    game_rush_2['Y/A']=round(game_rush_2['Yds']/game_rush_2['Att'],1)
    game_rush_2['Tm']=game_rush_2.iloc[:,1]
    game_rush_2['Opponent']=game_rush_2.iloc[:,11]
    game_rush_2=game_rush_2.iloc[:,[0,13,2,3,4,5,12,14]]
    game_rush_2['Year']=year
    return game_rush_2
```
